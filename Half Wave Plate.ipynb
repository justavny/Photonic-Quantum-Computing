# The Half Wave Plate

A half-wave plate (HWP) is an optical device made from a birefringent material that introduces a phase difference of π radians (180°) between two orthogonal polarization components of light. Due to this phase shift, the polarization state of incident light can be modified in a controlled manner. For linearly polarized light, a half-wave plate rotates the plane of polarization by an angle equal to twice the angle between the incident polarization direction and the optical axis of the wave plate. For circularly polarized light, it reverses the handedness of the polarization.  

In our KLM CNOT gate setup, the half-wave plate is oriented with its optical axis at 22.5° relative to the horizontal polarization direction, allowing it to implement the operation of a Hadamard gate on the target qubit. For example,  
Input State of Target: |H>  
Output State of Target: $\frac{1}{\sqrt{2}}$ (|H> +|V>)  
(assuming the polarization basis of photon to be {|H>, |V>})  
A CNOT gate can be decomposed into a sequence consisting of a Hadamard gate on the target qubit, followed by a controlled-phase (CZ) gate acting on both the control and target qubits, and finally another Hadamard gate on the target qubit. Consequently, half-wave plates are employed before and after the CZ operation to implement the required basis transformations.  

## Mathematical Representation of the Half-Wave Plate

A birefringent crystal possesses two principal axes, commonly referred to as the fast axis and the slow axis. Since light propagates through these axes with different refractive indices, a relative phase difference is accumulated between the corresponding polarization components.  

The phase retardation introduced by the crystal is  

$$
\delta = \frac{2\pi}{\lambda} |n_f - n_s| \times d
$$  

where:  
* ${n_f}$  is the refractive index along the fast axis,  
* ${n_s}$  is the refractive index along the slow axis,  
* d is the thickness of the crystal.  
 
For a Half-Wave Plate (HWP), the thickness is chosen such that the phase difference becomes  

$$
\delta = \pi.
$$

Therefore, in the ({|f>, |s>}) basis, the Jones matrix of an ideal HWP is  

HWP =
$$
\begin{bmatrix}
1 & 0 \\  
0 & -1
\end{bmatrix}.
$$  

The above matrix assumes that the incoming photon is already expressed in the fast-slow basis of the crystal.  

To describe a HWP whose optical axis is rotated by an angle (\theta), we introduce the rotation matrix  

$$
R(\theta)=
\begin{bmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{bmatrix}.
$$

The generalized HWP matrix is then given by  

$$
HWP(\theta) = 

R(-\theta)
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
R(\theta).
$$  

(Explanation of the order of multiplication: We first transform the polarization state into the fast-slow basis of the rotated crystal using R(${\theta}$), apply the phase retardation of the HWP, and then transform back to the original basis using R(−${\theta}$).)
Carrying out the matrix multiplication gives    

$$
HWP(\theta) = 

\begin{bmatrix}
\cos(2\theta) & -\sin(2\theta) \\
-\sin(2\theta) & -\cos(2\theta)
\end{bmatrix}.
$$  

In our KLM CNOT gate setup, the optical axis of the HWP is oriented at  

$$
\theta = 22.5^\circ = \frac{\pi}{8}.
$$

Substituting this value,  

$$
\cos(2\theta) = \cos\left(\frac{\pi}{4}\right) = 
\sin\left(\frac{\pi}{4}\right) = 
\frac{1}{\sqrt{2}}.
$$  

Hence the HWP matrix becomes  

$$
HWP\left(\frac{\pi}{8}\right) = 

\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 & -1 \\
-1 & -1
\end{bmatrix}.
$$  

Ignoring the physically irrelevant global phase factor of (-1), this matrix is equivalent to the Hadamard gate  

$$
H=
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 & 1 \\
1 & -1
\end{bmatrix}.
$$

Therefore, a Half-Wave Plate with its optical axis oriented at ${(22.5^\circ)}$ can be used to implement the Hadamard operation on the polarization qubit.  


Common materials include:  

Quartz (SiO₂) – the most widely used material in visible-light quantum optics experiments because of its excellent optical quality, low absorption, and stable birefringence.  
Calcite (CaCO₃) – exhibits very strong birefringence and is often used in polarizing beam splitters and polarization optics. The Nature 02054 setup specifically mentions calcite PBSs.  
Magnesium Fluoride (MgF₂) – commonly used for broadband wave plates.  
Mica – sometimes employed in multi-order wave plates due to ease of fabrication.
