# Beam Splitters (as in Quantum Optics)

A beam splitter can be modelled (simply) as a glass slab with a partially reflecting surface, so when a beam of light is incident upon it, it is partially transmitted and partially reflected.  
$|E_{in}|^2 = |E_r|^2 + |E_t|^2$ 
where   
E is the field amplitude, $E_{in}$ for incident light, $E_{r}$ for reflected light and $E_{t}$ for transmitted light.  
The above is obtained by using the fact that energy remains conserved.  

In Quantum optics, we replace the classical complex field amplitude E by annhilation opertor $\hat{a}$. It was observed that if the commutation relation of the annhilation operators does not allow us to obey the energy conservation theorem. This is because we only assumed one beam of light in the classical picture. However, in the quantum picture, even vacuum has some energy fluctuations, therefore we must add a term, let $\hat{a}_0$, to represent the vacuum state.  

On solving the two relations, we obtain:-  
$$
\begin{pmatrix}
\hat{a}_2 \\
\hat{a}_3
\end{pmatrix}
 = 
\begin{pmatrix}
t' & r \\
r' & t
\end{pmatrix}

\begin{pmatrix}
\hat{a}_0  \\
\hat{a}_1  
\end{pmatrix}
$$  

This program will demonstrate how to encode a Beam Splitter.

```python
# Using Strawberry Fields
import strawberryfields as sf
from strawberryfields import ops
from math import pi
# Create a 2-mode program
prog = sf.Program(2)
with prog.context as q:
    # Apply a beamsplitter to the two modes
    ops.BSgate(pi/4, 0) | (q[0], q[1])
prog.print()
prog.draw_circuit()

# Run the program on the Fock backend
eng = sf.Engine("fock", backend_options={"cutoff_dim": 5})  
result = eng.run(prog)
# Print the resulting state
print(result.state)
```

```python
# Using simple linear algebra with numpy
import numpy as np

# Define the beamsplitter unitary
theta = pi/4    
BS = np.array([[np.cos(theta), -np.sin(theta)],
               [np.sin(theta), np.cos(theta)]])

# Define the input state, let's say |1,0> (one photon in the first mode, zero in the second)
input_state = np.array([1, 0])

# Apply the beamsplitter to the input state
output_state = BS @ input_state

print(output_state)
```
